## 1. はじめての `print`

コンピュータに文字を表示してもらうには `print()` を使います。
表示したい文字は **ダブルクォート `"`** で囲みます。

```python
print("こんにちは、Python！")  # 文字を画面に表示する
```

```python
print("わたしの名前は ○○ です")  # ○○ を自分の名前に変えてみよう
```

`#` から後ろは **コメント** です。プログラムには影響せず、メモとして使えます。

## 2. Pythonは電卓になる

数の計算もできます。記号の意味はこちら。

| 記号 | 意味 | 例 |
|---|---|---|
| `+` | たし算 | `3 + 4` |
| `-` | ひき算 | `10 - 2` |
| `*` | かけ算 | `6 * 7` |
| `/` | わり算 | `10 / 4` |
| `//` | わり算の商（整数） | `10 // 4` |
| `%` | わり算のあまり | `10 % 4` |
| `**` | べき乗 | `2 ** 10` |

```python
3 + 4  # セルの最後の式の値は自動で表示される（ここでは 7）
```

> 💡 セルの**最後の行**の値は、`print` を書かなくても自動で表示されます。

```python
print(10 / 4)   # わり算は小数になる
print(10 // 4)  # 商
print(10 % 4)   # あまり
print(2 ** 10)  # 2の10乗
```

### 計算の順番
算数と同じで、かけ算・わり算が先。順番を変えたいときは `()` を使います。

```python
print(2 + 3 * 4)     # = 14
print((2 + 3) * 4)   # = 20
```

## 3. エラーはこわくない

わざと間違えてみましょう。下のセルを実行するとエラーが出ます。

```python
print("とじカッコを忘れた"  # わざとエラー：閉じカッコ ) を忘れている
```

上のセルを `print("とじカッコを忘れた")` に直して、もう一度実行してみよう。
