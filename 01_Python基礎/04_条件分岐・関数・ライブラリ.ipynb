## 1. if で場合分け

条件が `True` のときだけ実行されます。

```python
score = 72             # 点数
# 点数によって表示を変える（場合分け）
if score >= 80:
    print('すごい！')    # 80以上のとき
elif score >= 60:
    print('合格')        # 60以上80未満のとき
else:
    print('もう少し')    # それ未満のとき
```

### 比較の記号

| 記号 | 意味 |
|---|---|
| `==` | 等しい |
| `!=` | 等しくない |
| `>` `>=` | 大きい / 以上 |
| `<` `<=` | 小さい / 以下 |

`and`（かつ）、`or`（または）で組み合わせられます。

```python
temp = 28              # 気温
# 25以上 かつ 30未満 なら「夏日」
if temp >= 25 and temp < 30:
    print('夏日です')
```

## 2. 関数＝自分で作る命令

同じ処理を何度も使うときは関数にまとめます。

```python
# 名前を受け取ってあいさつ文を返す関数を定義
def greet(name):
    return f'こんにちは、{name}さん！'

print(greet('はなこ'))   # 関数を呼び出して結果を表示
print(greet('たろう'))
```

### 例：平均を返す関数

```python
# 数のリストを受け取って平均を返す関数
def average(numbers):
    return sum(numbers) / len(numbers)   # 合計÷個数

print(average([60, 70, 80]))   # → 70.0
print(average([100, 90]))      # → 95.0
```

## 3. ライブラリを import する

Pythonには便利な道具がたくさん付いてきます。`import` で呼び出します。

```python
import math               # 数学の便利機能を読み込む
print(math.sqrt(144))     # 平方根 → 12.0
print(math.pi)            # 円周率

import random             # 乱数の機能を読み込む
print(random.randint(1, 6))  # 1〜6のランダムな整数（サイコロ）
```
