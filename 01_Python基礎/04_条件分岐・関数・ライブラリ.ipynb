## 1. if で場合分け

条件が `True` のときだけ実行されます。

```python
score = 72
if score >= 80:
    print('すごい！')
elif score >= 60:
    print('合格')
else:
    print('もう少し')
```

### 比較の記号

| 記号 | 意味 |
|---|---|
| `==` | 等しい |
| `!=` | 等しくない |
| `>` `>=` | 大きい / 以上 |
| `<` `<=` | 小さい / 以下 |

`and`（かつ）、`or`（または）で組み合わせられます。

```python
temp = 28
if temp >= 25 and temp < 30:
    print('夏日です')
```

## 2. 関数＝自分で作る命令

同じ処理を何度も使うときは関数にまとめます。

```python
def greet(name):
    return f'こんにちは、{name}さん！'

print(greet('はなこ'))
print(greet('たろう'))
```

### 例：平均を返す関数

```python
def average(numbers):
    return sum(numbers) / len(numbers)

print(average([60, 70, 80]))
print(average([100, 90]))
```

## 3. ライブラリを import する

Pythonには便利な道具がたくさん付いてきます。`import` で呼び出します。

```python
import math
print(math.sqrt(144))   # 平方根 → 12.0
print(math.pi)          # 円周率

import random
print(random.randint(1, 6))  # サイコロ（1〜6のランダム）
```
