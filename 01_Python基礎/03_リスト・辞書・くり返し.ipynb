# 03. リスト・辞書・くり返し

**ゴール**
- たくさんの値をリストでまとめて扱える
- 辞書で「名前と値」を対応づけられる
- `for` で同じ処理をくり返せる

ここは統計の章でデータを扱う土台になります。

## 1. リスト＝値をならべた箱

`[ ]` の中にカンマ区切りでならべます。

```python
scores = [60, 75, 90, 55, 80]
print(scores)
print(scores[0])     # 1番目（番号は0から！）
print(scores[-1])    # 最後
print(len(scores))   # 個数
```

### リストの計算
合計・最大・最小・平均などがすぐ出せます。

```python
print(sum(scores))            # 合計
print(max(scores))            # 最大
print(min(scores))            # 最小
print(sum(scores) / len(scores))  # 平均
```

### リストに追加・並べ替え

```python
scores.append(100)   # 末尾に追加
print(scores)
scores.sort()        # 小さい順に並べ替え
print(scores)
```

## 2. 辞書＝名前と値のペア

`{キー: 値}` の形。名前（キー）で値を取り出せます。

```python
student = {"名前": "はなこ", "数学": 90, "英語": 75}
print(student["名前"])
print(student["数学"])
```

## 3. for でくり返す

リストの要素を1つずつ取り出して、同じ処理をします。

```python
for s in scores:
    print(s, '点')
```

> ⚠️ `for` の次の行は **4つの空白で字下げ（インデント）** します。Jupyterは自動でやってくれます。

### range でくり返し回数を決める

```python
for i in range(5):    # 0,1,2,3,4
    print(i, 'かいめ')
```

### 例：60点以上が何人いるか数える

```python
scores = [60, 75, 90, 55, 80, 100, 45]
count = 0
for s in scores:
    if s >= 60:
        count = count + 1
print('合格者は', count, '人')
```
