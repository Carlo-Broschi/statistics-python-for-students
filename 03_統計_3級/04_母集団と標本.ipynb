```python
import numpy as np               # 数値計算
import matplotlib.pyplot as plt  # グラフ描画
from scipy import stats          # 統計関数
plt.rcParams['axes.unicode_minus'] = False       # マイナス記号の文字化けを防ぐ
rng = np.random.default_rng(7)   # 乱数生成器（seedで結果を固定）
```

## 1. 母集団と標本

- **母集団**：本当に知りたい全体（例：日本の高校生全員の身長）
- **標本**：実際に調べた一部（例：100人の身長）
- **標本平均**から**母平均**を推測する

全数調査が難しいので、標本調査で代用します。

仮想の母集団（平均165, 標準偏差8の身長10万人）を作り、そこから標本を取ってみます。

```python
population = rng.normal(165, 8, 100000)    # 仮想の母集団（平均165・標準偏差8の身長10万人）
print('母平均:', population.mean().round(2))

sample = rng.choice(population, size=100)  # そこから100人を無作為に抽出
print('標本平均(100人):', sample.mean().round(2))
print('→ 全部調べなくても、母平均にかなり近い！')
```

## 2. 標本平均はばらつく（標本分布）

標本を取り直すたびに標本平均は少し変わります。
「標本平均を何度も計算した分布」を調べます。

```python
# 100人の標本平均を2000回くり返して集める
means = [rng.choice(population, 100).mean() for _ in range(2000)]
plt.hist(means, bins=40, edgecolor='white')         # 標本平均の分布
plt.axvline(165, color='red', linestyle='--', label='母平均')  # 母平均の位置
plt.title('標本平均の分布（n=100を2000回）'); plt.legend()
plt.show()
print('標本平均たちの平均:', np.mean(means).round(2))      # 母平均165に近い
print('標本平均たちの標準偏差:', np.std(means).round(3))    # ＝標準誤差
```

### 中心極限定理
標本平均の分布は、サンプルサイズ n が大きいほど**正規分布に近づき**、
その標準偏差（=標準誤差）は
$$ 標準誤差 = \frac{母標準偏差}{\sqrt{n}} $$
で小さくなります。「nを増やすほど推定が正確になる」根拠です。

```python
# サンプルサイズnを変えて標準誤差（母標準偏差÷√n）を比べる
for n in [25, 100, 400]:
    se = 8 / np.sqrt(n)                    # 標準誤差
    print(f'n={n:3d} → 標準誤差 {se:.3f}（nが4倍で誤差は半分）')
```

> ⚠️ **よくある間違い**：**標準偏差**（データ自体の散らばり）と**標準誤差**（標本平均の散らばり）は別物。標準誤差は n を増やすほど小さくなり、推定が正確になります。

## 📒 用語集 ＆ チートシート

| 用語 | 意味 |
|---|---|
| 母集団 | 本当に知りたい全体 |
| 標本 | 実際に調べた一部 |
| 標本平均 | 標本の平均（母平均の推定） |
| 標準誤差 | 標本平均の散らばり＝σ/√n |
| 中心極限定理 | nが大きいと標本平均は正規分布に近づく |
| 大数の法則 | nが大きいと実測が理論に近づく |

```python
# チートシート（標本）― そのまま実行できます
import numpy as np
sigma, n = 8, 100
se = sigma / np.sqrt(n)                 # 標準誤差 = σ/√n
print('標準誤差:', round(se, 3))
pop = np.random.default_rng(0).normal(165, 8, 10000)   # 母集団の例
sample = np.random.default_rng(1).choice(pop, 100)     # 標本を抽出
print('標本平均:', round(sample.mean(), 2))
```
