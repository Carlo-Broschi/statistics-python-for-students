```python
import numpy as np               # 数値計算
import matplotlib.pyplot as plt  # グラフ描画
from scipy import stats          # 統計関数（分布・検定など）
plt.rcParams['axes.unicode_minus'] = False       # マイナス記号の文字化けを防ぐ
```

## 1. 二項分布

「成功確率 p の試行を n 回くり返したときの成功回数」の分布。
例：コインを10回投げて表が出る回数。

```python
n, p = 10, 0.5                             # 試行回数nと成功確率p
k = np.arange(0, n + 1)                    # 成功回数の候補（0〜10）
prob = stats.binom.pmf(k, n, p)            # 各回数の確率（二項分布）

plt.bar(k, prob)                           # 棒グラフで分布を表示
plt.xlabel('表が出た回数'); plt.ylabel('確率')
plt.title(f'二項分布 B(n={n}, p={p})')
plt.show()
print('平均(np):', n*p, ' 標準偏差:', round(np.sqrt(n*p*(1-p)), 2))  # 平均np・標準偏差√(np(1-p))
```

```python
# 表がちょうど7回出る確率
print('P(X=7) =', round(stats.binom.pmf(7, 10, 0.5), 4))     # pmf＝ちょうどその回数の確率
# 表が8回以上出る確率
print('P(X>=8) =', round(1 - stats.binom.cdf(7, 10, 0.5), 4))  # 1−(7回以下の累積)
```

## 2. 正規分布

自然界・社会で最もよく現れる、左右対称のつりがね型。
身長・テストの点数などが近似的に従います。

> **数IIIメモ（ここだけ1分）**：正規分布の式には eˣ（指数関数）が出てきますが、**式の暗記は不要**です。押さえるのは「**連続データの確率＝曲線の下の“面積”**」という考え方だけ。その面積（確率）は `scipy` が計算してくれます。

```python
mu, sigma = 50, 10   # 平均50, 標準偏差10（偏差値の分布）
x = np.linspace(10, 90, 200)               # 横軸の値
y = stats.norm.pdf(x, mu, sigma)           # 正規分布の高さ（確率密度）
plt.plot(x, y)                             # 曲線を描く
plt.axvline(mu, color='red', linestyle='--', label='平均')  # 平均の位置に縦線
plt.title('正規分布 N(50, 10²)'); plt.legend()
plt.show()
```

## 3. 68-95-99.7 ルール

正規分布では、平均から
- ±1σ の範囲に約 **68%**
- ±2σ の範囲に約 **95%**
- ±3σ の範囲に約 **99.7%**
のデータが入ります。

```python
# 平均から±kσの範囲に入る割合を計算
for k in [1, 2, 3]:
    p = stats.norm.cdf(k) - stats.norm.cdf(-k)  # (+kσ以下の確率) − (−kσ以下の確率)
    print(f'±{k}σ の中に {p*100:.1f}%')
```

### 確率の計算（標準化を使う）
偏差値が70以上（平均50,σ10）の人は何%？

```python
# z = (70-50)/10 = 2.0 → 上側の確率
p = 1 - stats.norm.cdf(70, 50, 10)         # 70以上になる確率（上側）
print(f'偏差値70以上は上位 {p*100:.1f}%')
```

> **よくある間違い**：`pmf`（ちょうどその値の確率）と `cdf`（その値**以下**の累積）は別物。また正規分布など連続分布では「ちょうど◯◯」の確率は0で、**範囲（面積）**で考えます。

## 4. 「上位◯%のライン」を求める（パーセント点）

これまでは「点数 → 確率」を求めました。逆に「確率 → 点数」を求めるのが**パーセント点**（`ppf`：cdfの逆関数）です。上位◯%の境目や、信頼区間の限界値で使います。

```python
from scipy import stats
mu, sd = 50, 10
top5 = stats.norm.ppf(0.95, mu, sd)        # 上位5%の境目（=95パーセント点）
print(f'平均50・SD10のテストで上位5%に入るには {top5:.1f}点以上')
# 偏差値70 は z=2。上位何%にあたる？
print(f'偏差値70(z=2)は上位 {(1 - stats.norm.cdf(2))*100:.1f}%')
```

> `cdf`（点数→下側確率）と `ppf`（確率→点数）は逆の関係。「上位5%」「信頼係数95%」のような“割合から境目を逆算”する場面では `ppf` を使います。

## 用語集 ＆ チートシート

| 用語 | 意味 |
|---|---|
| 二項分布 | 成功/失敗をn回くり返した成功数の分布 |
| 正規分布 | 左右対称のつりがね型 |
| pmf | ちょうどその値の確率（離散） |
| pdf | 確率密度（連続）。確率は面積 |
| cdf | その値以下の累積確率 |
| 68-95-99.7 | ±1σ/±2σ/±3σ に入る割合 |

```python
# チートシート（確率分布）
stats.binom.pmf(k, n, p)        # 二項：ちょうどk回
stats.binom.cdf(k, n, p)        # 二項：k回以下
stats.norm.pdf(x, mu, sigma)    # 正規：高さ
stats.norm.cdf(x, mu, sigma)    # 正規：x以下の確率
1 - stats.norm.cdf(x, mu, sigma)  # x以上（上側）
```
